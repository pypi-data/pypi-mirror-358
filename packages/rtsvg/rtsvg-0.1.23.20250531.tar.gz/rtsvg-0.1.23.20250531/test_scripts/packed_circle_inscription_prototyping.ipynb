```python
import rtsvg
import random
from math import pi, sqrt, atan2
rt = rtsvg.RACETrack()
```

```python
_circles_ = []
for i in range(8): _circles_.append((0.0, 0.0, 3.0))
_cp_      = rtsvg.CirclePacker(rt, _circles_)
_cp_
```

```python
_border_circles_ = []
_packed_circles_ = _cp_.packedCircles()
for k, v in _cp_.fwd.items(): _border_circles_.append(_packed_circles_[v])
_x0_, _y0_, _x1_, _y1_ = _cp_.__packedCircleExtents__()
_inset_ = max(_x1_ - _x0_, _y1_ - _y0_) * 0.1
_svg_ = [f'<svg x="0" y="0" width="800" height="800" viewBox="{_x0_ - _inset_} {_y0_ - _inset_} {_x1_ - _x0_ + 2*_inset_} {_y1_ - _y0_ + 2*_inset_}" xmlns="http://www.w3.org/2000/svg">']
_svg_.append(f'<rect x="{_x0_ - _inset_}" y="{_y0_ - _inset_}" width="{_x1_ - _x0_ + 2*_inset_}" height="{_y1_ - _y0_ + 2*_inset_}" fill="#ffffff" />')
for i in range(len(_border_circles_)):
    _c_ = _border_circles_[i]
    _svg_.append(f'<circle cx="{_c_[0]}" cy="{_c_[1]}" r="{_c_[2]}" fill="none" stroke="#000000" stroke-width="0.1" />')
def toDeg(_rad_): return _rad_ * 180.0 / pi
_xcen_, _ycen_ = (_x0_ + _x1_) / 2.0, (_y0_ + _y1_) / 2.0
_svg_.append(f'<circle cx="{_xcen_}" cy="{_ycen_}" r="0.3" fill="#ff0000" stroke="none" />')
_sorter_ = []
_furthest_ = None
for i in range(len(_border_circles_)):
    _c_  = _border_circles_[i]
    _l_  = sqrt((_c_[0] - _xcen_)**2 + (_c_[1] - _ycen_)**2) + _c_[2]
    _uv_ = rt.unitVector(((_xcen_, _ycen_), (_c_[0], _c_[1])))
    _xouter_, _youter_ = _xcen_ + _l_ * _uv_[0], _ycen_ + _l_ * _uv_[1]
    _svg_.append(f'<line x1="{_xcen_}" y1="{_ycen_}" x2="{_xouter_}" y2="{_youter_}" stroke="#000000" stroke-width="0.05" />')
    if _furthest_ is None or _l_ > _furthest_: _furthest_ = _l_
    _sorter_.append((_l_, _xouter_, _youter_, atan2(_youter_ - _ycen_, _xouter_ - _xcen_), i))
    _svg_.append(rt.svgText(f'{toDeg(_sorter_[-1][3]):.1f}', _xouter_, _youter_-0.2, txt_h=1, anchor='middle'))
_svg_.append(f'<circle cx="{_xcen_}" cy="{_ycen_}" r="{_furthest_}" fill="none" stroke="#000000" stroke-width="0.1" />')

# Pick the three largest distances as long as they have sufficient angular distances
_sorter_.sort(reverse=True)
i0 = 0
i1 = None
i  = 1
while i1 is None and i < len(_sorter_):
    _angle_d_  = pi - abs(abs(_sorter_[i][3] - _sorter_[i0][3]) - pi)
    if _angle_d_ > pi/3.0: i1 = i
    i += 1
i2 = None
i  = i1 + 1
while i2 is None and i < len(_sorter_):
    _angle0_d_  = pi - abs(abs(_sorter_[i][3] - _sorter_[i0][3]) - pi)
    _angle1_d_  = pi - abs(abs(_sorter_[i][3] - _sorter_[i1][3]) - pi)
    if _angle0_d_ > pi/3.0 and _angle1_d_ > pi/3.0: i2 = i
    i += 1

_svg_.append(f'<circle cx="{_sorter_[i0][1]}" cy="{_sorter_[i0][2]}" r="0.3" fill="#ff0000" stroke="none" />')
_svg_.append(f'<circle cx="{_sorter_[i1][1]}" cy="{_sorter_[i1][2]}" r="0.3" fill="#ff0000" stroke="none" />')
_svg_.append(f'<circle cx="{_sorter_[i2][1]}" cy="{_sorter_[i2][2]}" r="0.3" fill="#ff0000" stroke="none" />')

# Construct a circle out of those three points
_circle_ = rt.threePointCircle((_sorter_[i0][1], _sorter_[i0][2]), (_sorter_[i1][1], _sorter_[i1][2]), (_sorter_[i2][1], _sorter_[i2][2]))
_svg_.append(f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="{_circle_[2]}" fill="none" stroke="#00ff00" stroke-width="0.1" />')
_svg_.append(f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="0.3" fill="#00ff00" stroke="none" />')

_svg_.append('</svg>')
rt.tile([''.join(_svg_)])
```

```python
_svg_ = [f'<svg x="0" y="0" width="300" height="300" viewBox="{_x0_ - _inset_} {_y0_ - _inset_} {_x1_ - _x0_ + 2*_inset_} {_y1_ - _y0_ + 2*_inset_}" xmlns="http://www.w3.org/2000/svg">']
_svg_.append(f'<rect x="{_x0_ - _inset_}" y="{_y0_ - _inset_}" width="{_x1_ - _x0_ + 2*_inset_}" height="{_y1_ - _y0_ + 2*_inset_}" fill="#ffffff" />')
_svg_hdr_ = ''.join(_svg_)

_svg_ = []
for i in range(len(_border_circles_)):
    _c_ = _border_circles_[i]
    _svg_.append(f'<circle cx="{_c_[0]}" cy="{_c_[1]}" r="{_c_[2]}" fill="none" stroke="#000000" stroke-width="0.1" />')
_svg_borders_ = ''.join(_svg_)

_svg_ftr_ = '</svg>'

_xy_            = None
_tiles_         = []
_all_positions_ = []
_all_positions_.append(f'<circle cx="{_xcen_}" cy="{_ycen_}" r="0.3" fill="#00ff00" stroke="none" />')

for i in range(14):
    _circle_, _points_ = _cp_.__approximateInscribedCircle__(xy=_xy_)
    _xy_ = (_circle_[0], _circle_[1])
    _all_positions_.append(f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="0.3" fill="#ff0000" stroke="none" />')
    _tiles_.append(_svg_hdr_ + _svg_borders_ + \
                   f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="0.3" fill="#ff0000" stroke="none" />' + \
                   f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="{_circle_[2]}" fill="none" stroke="#000000" stroke-width="0.1" />' + \
                   _svg_ftr_)
_tiles_.insert(0, _svg_hdr_ + _svg_borders_ + ''.join(_all_positions_) + _svg_ftr_)
rt.table(_tiles_, per_row=5, spacer=10)
```

```python
_circle_ = _cp_.optimalInscriptionCircle()
rt.tile([_svg_hdr_ + _svg_borders_ + \
         f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="0.3" fill="#ff0000" stroke="none" />' + \
         f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="{_circle_[2]}" fill="none" stroke="#000000" stroke-width="0.1" />' + \
         _svg_ftr_])
```

```python
into_circle = (10.0,15.0,2.0)
_border_circles_
_circle_

_x0_, _y0_, _x1_, _y1_ = -10.0, -10.0, 50.0, 30.0
_svg_ = [f'<svg x="0" y="0" width="800" height="600" viewBox="{_x0_} {_y0_} {_x1_ - _x0_} {_y1_ - _y0_}">']
_svg_.append(f'<rect x="{_x0_}" y="{_y0_}" width="{_x1_-_x0_}" height="{_y1_-_y0_}" fill="#ffffff" />')
_svg_.append(_svg_borders_)
_svg_.append(f'<circle cx="{_circle_[0]}" cy="{_circle_[1]}" r="{_circle_[2]}" fill="none" stroke="#000000" stroke-width="0.1" />')
_svg_.append(f'<circle cx="{into_circle[0]}" cy="{into_circle[1]}" r="{into_circle[2]}" fill="none" stroke="#ff0000" stroke-width="0.1" />')

scale = _circle_[2] / into_circle[2]
for i in range(len(_border_circles_)):
    _to_move_ = _border_circles_[i]
    x = (_to_move_[0] - _circle_[0])/scale + into_circle[0]
    y = (_to_move_[1] - _circle_[1])/scale + into_circle[1]
    r = _to_move_[2]/scale
    _svg_.append(f'<circle cx="{x}" cy="{y}" r="{r}" fill="none" stroke="#ff0000" stroke-width="0.1"/>')

_svg_.append('</svg>')
rt.tile([''.join(_svg_)])
```
