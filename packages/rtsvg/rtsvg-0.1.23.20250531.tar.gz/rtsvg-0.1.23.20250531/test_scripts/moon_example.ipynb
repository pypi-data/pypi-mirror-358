````python
import os
if os.path.exists('../../data/moon_example.parquet') == False:
    from ollama import chat, ChatResponse
    from pydantic import BaseModel
    class ContentUnitExcerpt(BaseModel):
        content_unit: str
        excerpt:      str
    def makePrompt(summary, summary_content_unit, content_unit):
        return f'''For the following summary, extract the the most relevant text that describes the content unit, "{content_unit}", from the following summary.  ''' + \
            f'''Use ellipses if necessary.  Return the results as a python String.\n\n"{summary}"'''
    model      = 'deepseek-r1:14b'
    _excerpts_ = []
    for i in range(len(df)):
        prompt = makePrompt(df.iloc[i]['summary'], df.iloc[i]['summary_content_unit'], df.iloc[i]['content_unit'])
        response: ChatResponse = chat(model=model, messages=[
            { 'role':     'user', 
            'content':  prompt,},],
            format=ContentUnitExcerpt.model_json_schema())
        contentUnitExcerpt = ContentUnitExcerpt.model_validate_json(response['message']['content'])
        _excerpts_.append(contentUnitExcerpt.excerpt)
    df['excerpt'] = _excerpts_
    df.to_parquet('../../data/moon_example.parquet')
df = pd.read_parquet('../../data/moon_example.parquet')
````
